```python
#detecting the outliers and store it 
#this array will store the outliers of each column
def detect_outliers_iqr(data):
    outliers = []
    data = sorted(data)
    q1 = np.percentile(data, 25)
    q3 = np.percentile(data, 75)
    IQR = q3-q1
    lwr_bound = q1-(1.5*IQR)
    upr_bound = q3+(1.5*IQR)
    for i in data: 
        if (i<lwr_bound or i>upr_bound):
            outliers.append(i)
    return outliers
```

# SECOND PHASE:Treating noise data

note: columns(longitude and housing_median_age) dont have any noise data ,while columns( median_house_value,median_income,total_rooms) have noise value AND WILL BE TREATED

```python
def treat_noise(column):
    tenth_percentile = np.percentile(column, 10)
    ninetieth_percentile = np.percentile(column, 90)
    print("Tenth_percintile: ",tenth_percentile," , ninetieth_percentile: ", ninetieth_percentile)
    x = np.where(column<tenth_percentile, tenth_percentile, column)
    x = np.where(x>ninetieth_percentile, ninetieth_percentile, x)
    return x
```

```python
TOTAL_ROOMS=treat_noise(data["total_rooms"])
TOTAL_ROOMS
#know total_rooms has no noise data
```

```python
#scatter plot for latitude and longitude columns 
plt.scatter(data.latitude,data.longitude)
plt.xlabel('longitude')
plt.ylabel('latitude')
```

```python
#elbow method for determining ptimal value for number of clusters
wss = []
k_rng = range(1,10)
for k in k_rng:
 km = KMeans(n_clusters=k)
 km.fit(features[['longitude','latitude']])
 wss.append(km.inertia_)
#'inertia' It is the sum of squared distances of samples to their closest cluster cente
plt.xlabel('K')
plt.ylabel('WSS')
plt.plot(k_rng, wss)
```

```python
#Create KMeans Classifier
km = KMeans(n_clusters=6, init = 'random', max_iter = 300, random_state = 0)
# number of clusters/centroids to create
# â€˜randomâ€™: choose n_clusters observations (rows) at random from
#'max_itier' this is the default value. This is the maximum number of iterations of the
#this is the default value. Determines random number generation 
#Use an int to make the randomness deterministic.
y_predicted = km.fit_predict(features[['longitude','latitude']])
#Run model on dataset
# method to compute the clusters and assign the labels
# fit_predict --> Compute cluster centers and predict cluster index for each sample.
```

```python
features['cluster']=y_predicted
features.head()
```

```python
#print the coordinates of each centroids
km.cluster_centers_
```

```python
#drawing clusters using seaborn
sns.relplot(x="longitude", y="latitude", hue="cluster", data=features, height=7)
plt.show()
```
